# **DATA** **VISUALIZATIONS**

First, let's create a pivot table and visualize the data in regard to various countries. The main core points concerning different countries can be easily visualized using the below pivot table.

```python
countrywise_pivot_table_make= pd.pivot_table(dataframe,values=["confirmed","deaths","recovered","active"],index="country",aggfunc=max)
#Im also using recovery rate and death rate to identify the intensity of it's rise.
countrywise_pivot_table_make["recovery rate"]=countrywise_pivot_table_make["recovered"]*100/countrywise_pivot_table_make["confirmed"]
countrywise_pivot_table_make["death rate/mortality rate"]=countrywise_pivot_table_make["deaths"]*100/countrywise_pivot_table_make["confirmed"]
countrywise_pivot_table_make = countrywise_pivot_table_make.sort_values(by="confirmed",ascending=False)
countrywise_pivot_table_make.style.background_gradient(cmap="cubehelix")
```

```python
plt.stackplot(x, confirmed, active, deaths, recovered, 
              labels = ['Confirmed', 'Active', 'Deaths', 'Recovered'])

plt.xlabel('Country', fontsize = '14')
plt.ylabel('No of Cases', fontsize = '14')
plt.xticks(x, rotation=300)
plt.ticklabel_format(style = 'plain', axis = 'y')
plt.title('Stackplot for same demonstration', fontsize = '18')
plt.legend(loc = 'upper right')
plt.grid(which='major', axis='both' , color = 'grey')

plt.show()
```

```python
def dotsize(n):
    temp = []
    for i in n:
        i /= 1e5
        temp.append(i)
    return temp

plt.scatter(x, confirmed, s=dotsize(confirmed), label = 'Confirmed')
plt.scatter(x, active, s=dotsize(active), label = 'Active')
plt.scatter(x, deaths, s=dotsize(deaths), label = 'Deaths')
plt.scatter(x, recovered, s=dotsize(recovered), label = 'Recovered')

plt.xlabel('Country', fontsize = '12')
plt.ylabel('No of Cases', fontsize = '12')
plt.xticks(x, x, rotation=300)
plt.ticklabel_format(style = 'plain', axis = 'y')
plt.title('Scatter Plot with varied size in dots to demonstrate more clearly', fontsize = '18')
plt.legend(loc = 'upper right')
plt.grid(which='major', axis='both')

plt.show()
```
